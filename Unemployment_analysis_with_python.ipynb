```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
data.columns = ["State","Date","Frequency","Estimated Unemployment Rate",
                "Estimated Employed", "Estimated Labour Participation Rate",
               "Region","Longitude","Latitude"]
```

```python
print(data)
```

```python
# Look at the correlation b/w the features of this dataset.
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(12,10))
sns.heatmap(data.corr())
plt.show()
```

```python
# look at the estimated number of employees according to different regions of india.
```

```python
data.columns = ["State","Date","Frequency","Estimated Unemployment Rate",
                "Estimated Employed", "Estimated Labour Participation Rate",
               "Region","Longitude","Latitude"]
plt.title("Indian Unemployment ")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()
```

```python
# see the unemployment rate according to different regions of india
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
```

```python
# Create a dashboard to analyze the unemployment rate of each indian state by region use sunburst plot.
```

```python
unemployment = data[["State","Region","Estimated Unemployment Rate"]]
figure = px.sunburst ( unemployment,path=["Region","State"],
                     values="Estimated Unemployment Rate",
                     width=700,height=700,color_continuous_scale="RdY1Gn",
                     title="Unemployment Rate in Indian")
figure.show()
```
